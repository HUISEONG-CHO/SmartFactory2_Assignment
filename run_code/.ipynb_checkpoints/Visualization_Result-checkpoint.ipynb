# (4) 시각화 및 결과 분석

## 1. Train_Size에 따른 F1 Score 비교

```python
print("F1_Macro 비교")
print("-----------------------------------------------")
print("Train_Size : ",500," / "," MFE-FNN : ",round(mfe_500_f1macro,4)," CNN : ",round(cnn_500_f1macro,4)," Stacking_Ensemble : ",round(se_500_f1macro,4))
print("Train_Size : ",5000," / "," MFE-FNN : ",round(mfe_5000_f1macro,4)," CNN : ",round(cnn_5000_f1macro,4)," Stacking_Ensemble : ",round(se_5000_f1macro,4))
print("Train_Size : ",50000," / "," MFE-FNN : ",round(mfe_50000_f1macro,4)," CNN : ",round(cnn_50000_f1macro,4)," Stacking_Ensemble : ",round(se_50000_f1macro,4))
print("Train_Size : ",162946," / "," MFE-FNN : ",round(mfe_162946_f1macro,4)," CNN : ",round(cnn_162946_f1macro,4)," Stacking_Ensemble : ",round(se_162946_f1macro,4))
```

```python
print("F1_Micro 비교")
print("-----------------------------------------------")
print("Train_Size : ",500," / "," MFE-FNN : ",round(mfe_500_f1micro,4)," CNN : ",round(cnn_500_f1micro,4)," Stacking_Ensemble : ",round(se_500_f1micro,4))
print("Train_Size : ",5000," / "," MFE-FNN : ",round(mfe_5000_f1micro,4)," CNN : ",round(cnn_5000_f1micro,4)," Stacking_Ensemble : ",round(se_5000_f1micro,4))
print("Train_Size : ",50000," / "," MFE-FNN : ",round(mfe_50000_f1micro,4)," CNN : ",round(cnn_50000_f1micro,4)," Stacking_Ensemble : ",round(se_50000_f1micro,4))
print("Train_Size : ",162946," / "," MFE-FNN : ",round(mfe_162946_f1micro,4)," CNN : ",round(cnn_162946_f1micro,4)," Stacking_Ensemble : ",round(se_162946_f1micro,4))
```

```python
# 그래프 그리기 전, 배열 생성
mfe_f1_macro=[round(mfe_500_f1macro,4),round(mfe_5000_f1macro,4),round(mfe_50000_f1macro,4),round(mfe_162946_f1macro,4)]
mfe_f1_micro=[round(mfe_500_f1micro,4),round(mfe_5000_f1micro,4),round(mfe_50000_f1micro,4),round(mfe_162946_f1micro,4)]

cnn_f1_macro=[round(cnn_500_f1macro,4),round(cnn_5000_f1macro,4),round(cnn_50000_f1macro,4),round(cnn_162946_f1macro,4)]
cnn_f1_micro=[round(cnn_500_f1micro,4),round(cnn_5000_f1micro,4),round(cnn_50000_f1micro,4),round(cnn_162946_f1micro,4)]

se_f1_macro=[round(se_500_f1macro,4),round(se_5000_f1macro,4),round(se_50000_f1macro,4),round(se_162946_f1macro,4)]
se_f1_micro=[round(se_500_f1micro,4),round(se_5000_f1micro,4),round(se_50000_f1micro,4),round(se_162946_f1micro,4)]

mfe_f1_macro=np.array(mfe_f1_macro)
mfe_f1_micro=np.array(mfe_f1_micro)

cnn_f1_macro=np.array(cnn_f1_macro)
cnn_f1_micro=np.array(cnn_f1_micro)

se_f1_macro=np.array(se_f1_macro)
se_f1_micro=np.array(se_f1_micro)
```

```python
x = np.array(["500","5000","50000","162946"])

plt.plot(x,mfe_f1_macro,'b',label='MFE-FNN')
plt.plot(x,cnn_f1_macro,'r',label='CNN')
plt.plot(x,se_f1_macro,'g',label='Stacking Ensemble')
plt.xlabel('Train_Data_Size')
plt.ylabel('F1 Score')
plt.title('F1_Macro')
plt.legend(loc='upper right')
plt.show()
```

```python
x = np.array(["500","5000","50000","162946"])

plt.plot(x,mfe_f1_micro,'b',label='MFE-FNN')
plt.plot(x,cnn_f1_micro,'r',label='CNN')
plt.plot(x,se_f1_micro,'g',label='Stacking Ensemble')
plt.xlabel('Train_Data_Size')
plt.ylabel('F1 Score')
plt.title('F1_Micro')
plt.legend(loc='upper right')
plt.show()
```
